```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

creating CSV dataset thanks to GPT4-o

```python
# ساختن داده‌ها به صورت CSV
data = """Patient_ID,Sore_Throat,Fever,Swollen_Glands,Congestion,Headache,Diagnosis
1,Yes,Yes,Yes,Yes,Yes,Strep throat
2,No,No,No,Yes,Yes,Allergy
3,Yes,Yes,No,No,No,Cold
4,Yes,No,Yes,No,No,Strep throat
5,Yes,No,No,Yes,Yes,Cold
6,No,No,Yes,No,No,Allergy
7,Yes,Yes,Yes,No,Yes,Strep throat
8,No,No,No,Yes,Yes,Allergy
9,Yes,Yes,No,No,Yes,Cold
10,Yes,Yes,No,Yes,Yes,Cold
"""

# ذخیره کردن داده‌ها در یک فایل CSV
with open('patients.csv', 'w') as file:
    file.write(data)

# خواندن فایل CSV با pandas
df = pd.read_csv('patients.csv')

# نمایش اطلاعات DataFrame
print(df)
```

```python
print('---df[0:5]')
print(df[0:5])
print('---df.size')
print(df.size)
print('---df.shape')
df.shape
```

```python
df = df.replace({"Yes":1, "No":0})
```

```python
df.drop(columns=["Patient_ID"])
```

```python
target = df['Diagnosis']
target
```

```python
target.replace({'Strep throat': 4, 'Allergy': 5, 'Cold': 6})
```

```python
new_df = df.drop(columns=["Diagnosis", "Patient_ID"])
```

```python
patientTree = DecisionTreeClassifier()
patientTree
```

```python
patientTree.fit(new_df, target)
```

visualization thanks to GPT 3.5

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from io import StringIO
import pydotplus
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Assuming new_df is your feature DataFrame and target is your target variable
# Train a decision tree model
patientTree = DecisionTreeClassifier()
patientTree.fit(new_df, target)

# Visualize the decision tree
dot_data = StringIO()
filename = "patient.png"
featureNames = new_df.columns[0:5]
out = export_graphviz(patientTree, feature_names=featureNames, out_file=dot_data, class_names=np.unique(target), filled=True, special_characters=True, rotate=False)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png(filename)
img = mpimg.imread(filename)
plt.figure(figsize=(100, 200))
plt.imshow(img, interpolation='nearest')
```
